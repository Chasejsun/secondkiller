```python
import os
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
import datetime
import time
import random
from threading import Thread

READY = False

def logged_in():
    try:
        driver.find_element_by_link_text("亲，请登录")
        return False
    except Exception as e:
        return True

def keep_login_and_wait(driver, buy_time, link):
    currentTime = datetime.datetime.now()
    if buy_time < currentTime:
        return
    
    while True:
        currentTime = datetime.datetime.now()
        if (buy_time - currentTime).seconds > 61:
            driver.get(link)
            print("刷新购物车界面，防止登录超时...")
            time.sleep(60)
        else:
            print("抢购时间点将近，停止自动刷新，准备进入抢购阶段...")
            break

    while (buy_time - currentTime).seconds > 3:
        1

def buy(driver, link):
    while True:
        try:
            driver.get(link)
            while not driver.find_element_by_id('J_LinkBuy'):
                1
#             driver.find_element_by_id('J_LinkBuy').click()
            driver.execute_script("document.getElementById('J_LinkBuy').click();")
            break
        except Exception as e:
            print(e)
    while True:
        try:
            while not driver.find_element_by_class_name('go-btn'):
                1
#             driver.find_element_by_class_name('go-btn').click()
            driver.execute_script("document.getElementById('go-btn').click();")
            print("已经点击提交订单按钮")
            break
        except Exception as e:
            print(e)
            
def login(profile):
    option = webdriver.ChromeOptions()
    option.add_argument("user-data-dir={}".format(profile))
    option.add_argument('disable-infobars')
    option.add_argument("--remote-debugging-port={}".format(random.randint(10000, 65535)))
    driver = webdriver.Chrome(options=option)
    driver.maximize_window()
    driver.get("https://www.taobao.com")
    try:
        driver.find_element_by_link_text("亲，请登录").click()
    except Exception as e:
        pass
    while not READY:
        1
    driver.quit()
    print('login completed')

def kill(profile, buy_time, link):
    #让浏览器不要显示当前受自动化测试工具控制的提醒
    option = webdriver.ChromeOptions()
    print("正在打开chrome浏览器...")
    option.add_argument("user-data-dir={}".format(profile))
#     option.add_argument('disable-infobars')
#     option.add_argument('--no-sandbox')
#     option.add_argument('--disable-dev-shm-usage')
    option.add_argument('--disable-gpu')
    option.add_argument('window-size=1920x1080')
    option.add_argument('--headless')
    prefs = {"profile.managed_default_content_settings.images": 2}
    option.add_experimental_option("prefs", prefs)
    option.add_argument("--remote-debugging-port={}".format(random.randint(10000, 65535)))
    driver = webdriver.Chrome(options=option)
    driver.maximize_window()
    print("chrome浏览器已经打开...")
    
    driver.get("https://www.taobao.com")
    assert logged_in()
    
    bt = datetime.datetime.strptime(buy_time, '%Y-%m-%d %H:%M:%S')
    
    keep_login_and_wait(driver, bt, link)
    buy(driver, link)
    driver.quit()

def run(profile):
    login(profile)
    kill(profile)
```

```python
for i in range(1):
    time.sleep(random.random())
    p = "selenium-{}".format(i)
    Thread(target=login, args=(p,)).start()
```
